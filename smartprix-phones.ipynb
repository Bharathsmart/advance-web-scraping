```python
import numpy as np
import pandas as pd
```

```python
name = []
price = []
specs = []

for i in containers:
    name.append(i.find("h2").text )
    price.append(i.find('span', { 'class': 'price'}).text)
    try:
        div = i.find('div', class_='score rank-1-bg') or i.find('div', class_='score rank-2-bg')
        if div and div.find('b'):
            specs.append(div.find('b').text.strip())
        else:
            specs.append(np.nan)
    except Exception as e:
        specs.append(np.nan)
        print(f"Error occurred: {e}")

```

```python
names = []
prices = []
rating = []
sim = []
processor = []
ram = []
battery = []
display = []
camera = []
card = []
os = []

for i in soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'}):
    try:
        names.append(i.find('h2').text)
    except:
        names.append(np.nan)
    try:
        prices.append(i.find('span',{'class':'price'}).text)
    except:
        price.append(np.nan)
    try:
        div = i.find('div', class_='score rank-1-bg') or i.find('div', class_='score rank-2-bg')
        if div and div.find('b'):
            rating.append(div.find('b').text.strip())
        else:
            rating.append(np.nan)
    except Exception as e:
        rating.append(np.nan)
        print(f"Error occurred: {e}")
        
    x = i.find('ul',{'class':'sm-feat specs'}).find_all('li')
    try:
        sim.append(x[0].text)
    except:
        sim.append(np.nan)
    try:
        processor.append(x[1].text)
    except:
        processor.append(np.nan)
    try:    
        ram.append(x[2].text)
    except:
        ram.append(np.nan)
    try:
        battery.append(x[3].text)
    except:
        battery.append(np.nan)
    try:
        display.append(x[4].text)
    except:
        display.append(np.nan)
    try:
        camera.append(x[5].text)
    except:
        camera.append(np.nan)
    try:
        card.append(x[6].text)
    except:
        card.append(np.nan)
    try:
        os.append(x[7].text)
    except:
        os.append(np.nan)
```
